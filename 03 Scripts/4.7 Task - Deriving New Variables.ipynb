# Question 2. Busiest Day to Busiest Days. Label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, theyâ€™d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
#Check for Busiest Days of the Year
df_ords_prods_merged['order_dow'].value_counts(dropna = False)
```

```python
#creating new if statement for loop for the new requirements
result2 = []

for value in df_ords_prods_merged['order_dow']:
  if value == 0 or value ==1:
    result2.append('Busiest days')
  elif value == 4 or value == 3:
    result2.append('Least busy days')
  else:
    result2.append('Regularly busy days')
```

```python
#Creating New Column 'Busiest Days'
df_ords_prods_merged['Busiest days'] = result2
```

```python
df_ords_prods_merged.head()
```

# Question 3. Check the values of this new column

```python
#Checking values of 'Busiest days' column
df_ords_prods_merged['Busiest days'].value_counts(dropna = False)
```

The loop is accurate because the Busiest days column is equal to the sum of the value 0 (Saturday) and value 1 (Sunday). The same is also true for the least busy days column. It matches with the sum of value 3 and value 4 which are the lowest values of any days of the week. 

# Question 4. Creating Busiest Period of Day Column

```python
#Check frequency of values in the order_hour_of_day column
df_ords_prods_merged['order_hour_of_day'].value_counts()
```

```python
#Create a for-loop showing values'originally displayed as order_hour_of_day to instead show periods of time
hour_of_day = []


for value in df_ords_prods_merged['order_hour_of_day']:
    if value in [10 ,11, 14, 15, 13, 12, 16, 9]:
     hour_of_day.append('Most_orders')
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
         hour_of_day.append('Fewest_orders')
    else: 
         hour_of_day.append('Average_orders')
```

```python
#Create new column called 'busiest_period_of_day'
df_ords_prods_merged['busiest_period_of_day'] = hour_of_day
```
