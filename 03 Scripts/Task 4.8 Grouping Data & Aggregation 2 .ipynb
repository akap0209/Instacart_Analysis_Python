# Question 6. Create a spending flag for each user based on the average price across all their orders using the following criteria: If the mean of the prices of products purchased by a user is lower than 10, then flag them as a â€œLow spender.â€ If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a â€œHigh spender.â€

```python
#Correction: Finding np.mean rather than np.max
ords_prods_merged['average_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#create low spender flag
ords_prods_merged.loc[ords_prods_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#create high spender flag
ords_prods_merged.loc[ords_prods_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merged.head()
```

```python
#Frequency check
ords_prods_merged['spending_flag'].value_counts(dropna= False)
```

# Question 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column. The criteria for the flag should be as follows: If the median of â€œdays_since_prior_orderâ€ is higher than 20, then the customer should be labeled a â€œNon-frequent customer.â€ If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a â€œRegular customer.â€ If the median is lower than or equal to 10, then the customer should be labeled a â€œFrequent customer.â€

```python
#Creating median_prior_orders column using transform()
ords_prods_merged['median_prior_orders'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged.head()
```

```python
#Creating order frequency flag 'Frequent Customer'
ords_prods_merged.loc[ords_prods_merged['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
#Creating order frequency flag 'Regular customer'
ords_prods_merged.loc[(ords_prods_merged['median_prior_orders'] >10) & (ords_prods_merged['median_prior_orders'] <=20), 'order_frequency_flag'] = 'Regular customer'
```

```python
#Creating order frequency flag 'Frequent Customer' 
ords_prods_merged.loc[ords_prods_merged['median_prior_orders'] <=10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merged.head()
```

```python
# Check the frequency

ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
